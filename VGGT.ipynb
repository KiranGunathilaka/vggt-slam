```python
import numpy as np, plotly.graph_objects as go, pycolmap
from pathlib import Path

# ---------- settings ----------
SCENE_DIR  = Path("/content/scene_subset")
COLOR_MODE = "rgb"   # "rgb" or "z"
POINT_SIZE = 1.0
POINT_OPACITY = 0.95
# ------------------------------

recon = pycolmap.Reconstruction(SCENE_DIR)

# Points + colors
pts, cols = [], []
for p in recon.points3D.values():
    pts.append(np.asarray(p.xyz, dtype=np.float64))
    c = getattr(p, "color", None)
    if c is None:
        c = getattr(p, "rgb", None)
    if c is None:
        c = np.array([200, 200, 200], dtype=np.uint8)  # default gray
    else:
        c = np.asarray(c, dtype=np.uint8)
    cols.append(c / 255.0)

pts  = np.asarray(pts, dtype=np.float64)
cols = np.asarray(cols, dtype=np.float64)

# Drop NaN/Inf only if present
bad = ~np.isfinite(pts).all(axis=1)
if bad.any():
    pts, cols = pts[~bad], cols[~bad]

print(f"Rendering points: {len(pts)} (expected total: {len(recon.points3D)})")

def qvec2rotmat(q):
    qw,qx,qy,qz = q
    return np.array([
        [1-2*(qy*qy+qz*qz), 2*(qx*qy - qz*qw), 2*(qx*qz + qy*qw)],
        [2*(qx*qy + qz*qw), 1-2*(qx*qx+qz*qz), 2*(qy*qz - qx*qw)],
        [2*(qx*qz - qy*qw), 2*(qy*qz + qx*qw), 1-2*(qx*qx+qy*qy)]
    ], dtype=float)

def get_Rcw_tcw(img):
    cm = img.cam_from_world
    if hasattr(cm, "matrix"):
        M = np.array(cm.matrix())
        if M.ndim==2 and M.shape[0]>=3 and M.shape[1]>=4:
            return M[:3,:3], M[:3,3]
    if hasattr(img, "qvec") and hasattr(img, "tvec"):
        return qvec2rotmat(np.array(img.qvec)), np.array(img.tvec)
    return np.eye(3), np.zeros(3)

def camera_center_world(Rcw, tcw):
    Rwc = Rcw.T
    Cw  = -Rwc @ tcw
    return Cw, Rwc

def frustum_lines(pycam, scale=0.05, z=1.0):
    K = pycam.calibration_matrix()
    fx, fy, cx, cy = K[0,0], K[1,1], K[0,2], K[1,2]
    w, h = pycam.width, pycam.height
    corners = np.array([[0,0],[w-1,0],[w-1,h-1],[0,h-1]], dtype=float)
    rays = []
    for x,y in corners:
        X = (x-cx)/fx * z; Y = (y-cy)/fy * z; Z = z
        rays.append([X,Y,Z])
    rays = np.array(rays)
    origin = np.zeros((1,3))
    pts_cam = np.vstack([origin, rays]) * scale
    lines = np.array([[0,1],[0,2],[0,3],[0,4],[1,2],[2,3],[3,4],[4,1]])
    return pts_cam, lines

# Scene bounds (NumPy 2.x: use np.ptp)
xyz_min, xyz_max = pts.min(axis=0), pts.max(axis=0)
center = (xyz_min + xyz_max) / 2
size   = float(np.ptp(xyz_max - xyz_min)) or 1.0
pad    = 0.05 * size

fig = go.Figure()

# Points
if COLOR_MODE.lower() == "z":
    z = pts[:,2]
    z_norm = (z - z.min()) / max(z.ptp(), 1e-6)
    fig.add_trace(go.Scatter3d(
        x=pts[:,0], y=pts[:,1], z=pts[:,2],
        mode='markers',
        marker=dict(size=POINT_SIZE, color=z_norm, colorscale='Viridis', opacity=POINT_OPACITY),
        name="points", hoverinfo='skip'
    ))
else:
    fig.add_trace(go.Scatter3d(
        x=pts[:,0], y=pts[:,1], z=pts[:,2],
        mode='markers',
        marker=dict(size=POINT_SIZE, color=cols, opacity=POINT_OPACITY),
        name="points", hoverinfo='skip'
    ))

# Camera centers
camX, camY, camZ = [], [], []
for img in recon.images.values():
    Rcw, tcw = get_Rcw_tcw(img)
    Cw, _ = camera_center_world(Rcw, tcw)
    camX.append(Cw[0]); camY.append(Cw[1]); camZ.append(Cw[2])

fig.add_trace(go.Scatter3d(
    x=camX, y=camY, z=camZ, mode='markers',
    marker=dict(size=2, color='red'),
    name='cameras', hoverinfo='skip'
))

# Frustums
frustum_scale = 0.04 * size
for img in recon.images.values():
    Rcw, tcw = get_Rcw_tcw(img)
    Cw, Rwc  = camera_center_world(Rcw, tcw)
    pycam    = recon.cameras[img.camera_id]
    fr_cam, fr_lines = frustum_lines(pycam, scale=frustum_scale, z=1.0)
    fr_world = (Rwc @ fr_cam.T).T + Cw
    for a,b in fr_lines:
        fig.add_trace(go.Scatter3d(
            x=[fr_world[a,0], fr_world[b,0]],
            y=[fr_world[a,1], fr_world[b,1]],
            z=[fr_world[a,2], fr_world[b,2]],
            mode='lines',
            line=dict(width=0.8, color='red'),
            showlegend=False, hoverinfo='skip'
        ))

fig.update_layout(
    scene=dict(
        xaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.08)'),
        yaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.08)'),
        zaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.08)'),
        xaxis_title="X", yaxis_title="Y", zaxis_title="Z",
        aspectmode='data'
    ),
    width=980, height=740,
    margin=dict(l=0, r=0, t=40, b=0),
    title="VGGT / COLMAP — full point cloud + thin markers",
    scene_camera=dict(eye=dict(x=1.4, y=1.2, z=1.0))
)

fig.update_scenes(
    xaxis_range=[center[0]-size/2 - pad, center[0]+size/2 + pad],
    yaxis_range=[center[1]-size/2 - pad, center[1]+size/2 + pad],
    zaxis_range=[center[2]-size/2 - pad, center[2]+size/2 + pad],
)

fig.show()
```

```python
import numpy as np, plotly.graph_objects as go, pycolmap
from pathlib import Path

# ---------- settings ----------
SCENE_DIR  = Path("/content/vggt/examples/single_cartoon/sparse")
COLOR_MODE = "rgb"   # "rgb" or "z"
POINT_SIZE = 1.0
POINT_OPACITY = 0.95
# ------------------------------

recon = pycolmap.Reconstruction(SCENE_DIR)

# Gather points + colors safely
pts_list, cols_list = [], []
for p in recon.points3D.values():
    xyz = np.asarray(p.xyz, dtype=np.float64).reshape(-1)  # ensure 1D length-3
    if xyz.shape[0] != 3 or not np.isfinite(xyz).all():
        continue
    c = getattr(p, "color", None)
    if c is None:
        c = getattr(p, "rgb", None)
    if c is None:
        c = np.array([200, 200, 200], dtype=np.uint8)
    c = np.asarray(c, dtype=np.uint8).reshape(-1)
    if c.shape[0] != 3:
        c = np.array([200, 200, 200], dtype=np.uint8)
    pts_list.append(xyz)
    cols_list.append(c / 255.0)

if len(pts_list) == 0:
    print("No valid 3D points in this reconstruction (or all were NaN/Inf).")
    # Still show cameras for debugging
    fig = go.Figure()
    camX, camY, camZ = [], [], []
    def q2R(q):
        qw,qx,qy,qz = q
        return np.array([
            [1-2*(qy*qy+qz*qz), 2*(qx*qy - qz*qw), 2*(qx*qz + qy*qw)],
            [2*(qx*qy + qz*qw), 1-2*(qx*qx+qz*qz), 2*(qy*qz - qx*qw)],
            [2*(qx*qz - qy*qw), 2*(qy*qz + qx*qw), 1-2*(qx*qx+qy*qy)]
        ], dtype=float)
    def get_Rcw_tcw(img):
        cm = img.cam_from_world
        if hasattr(cm, "matrix"):
            M = np.array(cm.matrix())
            if M.ndim==2 and M.shape[0]>=3 and M.shape[1]>=4:
                return M[:3,:3], M[:3,3]
        if hasattr(img, "qvec") and hasattr(img, "tvec"):
            return q2R(np.array(img.qvec)), np.array(img.tvec)
        return np.eye(3), np.zeros(3)
    for img in recon.images.values():
        Rcw, tcw = get_Rcw_tcw(img); Rwc = Rcw.T; Cw = -Rwc @ tcw
        camX.append(Cw[0]); camY.append(Cw[1]); camZ.append(Cw[2])
    if len(camX):
        fig.add_trace(go.Scatter3d(x=camX,y=camY,z=camZ,mode='markers',
                                   marker=dict(size=3,color='red'), name='cameras'))
    fig.update_layout(width=800,height=600, title="No points — camera centers only")
    fig.show()
else:
    pts  = np.vstack(pts_list)          # (N,3)
    cols = np.vstack(cols_list)         # (N,3)

    print(f"Rendering points: {len(pts)} (expected total: {len(recon.points3D)})")

    # Scene bounds (NumPy 2.x safe)
    xyz_min, xyz_max = pts.min(axis=0), pts.max(axis=0)
    center = (xyz_min + xyz_max) / 2
    size   = float(np.max(xyz_max - xyz_min)) or 1.0
    pad    = 0.05 * size

    # Helpers
    def qvec2rotmat(q):
        qw,qx,qy,qz = q
        return np.array([
            [1-2*(qy*qy+qz*qz), 2*(qx*qy - qz*qw), 2*(qx*qz + qy*qw)],
            [2*(qx*qy + qz*qw), 1-2*(qx*qx+qz*qz), 2*(qy*qz - qx*qw)],
            [2*(qx*qz - qy*qw), 2*(qy*qz + qx*qw), 1-2*(qx*qx+qy*qy)]
        ], dtype=float)
    def get_Rcw_tcw(img):
        cm = img.cam_from_world
        if hasattr(cm, "matrix"):
            M = np.array(cm.matrix())
            if M.ndim==2 and M.shape[0]>=3 and M.shape[1]>=4:
                return M[:3,:3], M[:3,3]
        if hasattr(img, "qvec") and hasattr(img, "tvec"):
            return qvec2rotmat(np.array(img.qvec)), np.array(img.tvec)
        return np.eye(3), np.zeros(3)
    def camera_center_world(Rcw, tcw):
        Rwc = Rcw.T
        Cw  = -Rwc @ tcw
        return Cw, Rwc
    def frustum_lines(pycam, scale=0.05, z=1.0):
        K = pycam.calibration_matrix()
        fx, fy, cx, cy = K[0,0], K[1,1], K[0,2], K[1,2]
        w, h = pycam.width, pycam.height
        corners = np.array([[0,0],[w-1,0],[w-1,h-1],[0,h-1]], dtype=float)
        rays = []
        for x,y in corners:
            X = (x-cx)/fx * z; Y = (y-cy)/fy * z; Z = z
            rays.append([X,Y,Z])
        rays = np.array(rays)
        origin = np.zeros((1,3))
        pts_cam = np.vstack([origin, rays]) * scale
        lines = np.array([[0,1],[0,2],[0,3],[0,4],[1,2],[2,3],[3,4],[4,1]])
        return pts_cam, lines

    fig = go.Figure()

    # Points
    if COLOR_MODE.lower() == "z":
        z = pts[:,2]
        z_norm = (z - z.min()) / max(np.ptp(z), 1e-6)
        fig.add_trace(go.Scatter3d(
            x=pts[:,0], y=pts[:,1], z=pts[:,2],
            mode='markers',
            marker=dict(size=POINT_SIZE, color=z_norm, colorscale='Viridis', opacity=POINT_OPACITY),
            name="points", hoverinfo='skip'
        ))
    else:
        fig.add_trace(go.Scatter3d(
            x=pts[:,0], y=pts[:,1], z=pts[:,2],
            mode='markers',
            marker=dict(size=POINT_SIZE, color=cols, opacity=POINT_OPACITY),
            name="points", hoverinfo='skip'
        ))

    # Cameras + frustums
    camX, camY, camZ = [], [], []
    for img in recon.images.values():
        Rcw, tcw = get_Rcw_tcw(img)
        Cw, Rwc  = camera_center_world(Rcw, tcw)
        camX.append(Cw[0]); camY.append(Cw[1]); camZ.append(Cw[2])
    if len(camX):
        fig.add_trace(go.Scatter3d(
            x=camX, y=camY, z=camZ, mode='markers',
            marker=dict(size=2, color='red'), name='cameras', hoverinfo='skip'
        ))
        frustum_scale = 0.04 * size
        for img in recon.images.values():
            Rcw, tcw = get_Rcw_tcw(img)
            Cw, Rwc  = camera_center_world(Rcw, tcw)
            pycam    = recon.cameras[img.camera_id]
            fr_cam, fr_lines = frustum_lines(pycam, scale=frustum_scale, z=1.0)
            fr_world = (Rwc @ fr_cam.T).T + Cw
            for a,b in fr_lines:
                fig.add_trace(go.Scatter3d(
                    x=[fr_world[a,0], fr_world[b,0]],
                    y=[fr_world[a,1], fr_world[b,1]],
                    z=[fr_world[a,2], fr_world[b,2]],
                    mode='lines',
                    line=dict(width=0.8, color='red'),
                    showlegend=False, hoverinfo='skip'
                ))

    fig.update_layout(
        scene=dict(
            xaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.08)'),
            yaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.08)'),
            zaxis=dict(showgrid=True, gridcolor='rgba(0,0,0,0.08)'),
            xaxis_title="X", yaxis_title="Y", zaxis_title="Z",
            aspectmode='data'
        ),
        width=980, height=740,
        margin=dict(l=0, r=0, t=40, b=0),
        title="VGGT / COLMAP — point cloud + cameras (robust)",
        scene_camera=dict(eye=dict(x=1.4, y=1.2, z=1.0))
    )

    fig.update_scenes(
        xaxis_range=[center[0]-size/2 - pad, center[0]+size/2 + pad],
        yaxis_range=[center[1]-size/2 - pad, center[1]+size/2 + pad],
        zaxis_range=[center[2]-size/2 - pad, center[2]+size/2 + pad],
    )

    fig.show()
```

```python
from pathlib import Path
import os

ROOT = Path("/content/vggt/examples/single_cartoon/sparse")  # <-- your path

def model_stats(model_dir: Path):
    files = ["cameras.bin","images.bin","points3D.bin","points3D.ply"]
    out = {}
    for f in files:
        p = model_dir / f
        out[f] = (p.exists(), os.path.getsize(p) if p.exists() else 0)
    return out

# Print stats for ROOT and any subfolders (e.g., sparse/0, sparse/1)
print("== ROOT ==", ROOT)
print(model_stats(ROOT))
for sub in sorted(ROOT.glob("*")):
    if sub.is_dir():
        print("== SUBDIR ==", sub)
        print(model_stats(sub))
```
